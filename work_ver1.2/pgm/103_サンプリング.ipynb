# 基本設定

```python
import setting

import pandas as pd
import numpy as np
import math
import pathlib as Path

import scipy as sc
import sklearn
import pickle
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sb
sb.set(font='IPAexGothic')

# データ格納先設定
data_path = Path.Path('../data')
# 接頭辞
prefix = 'ana103'
```

# データ抽出

```python
# 軸データ
base = pd.read_pickle(data_path / 'ana102_base.pkl')
base
```

# サンプリング
> 分割ルール
>> 訓練データ：テストデータ = 7:3

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(base, test_size=0.3, random_state=123, stratify=base['flg_sts'])
```

```python
# 分割フラグを付与
_train = df_train.copy()
_test  = df_test.copy()

_train['flg_train'] = 1
_test['flg_train'] = 0
```

```python
# 訓練データとテストデータを縦結合
base_smpl = pd.concat([_train, _test], axis=0)
base_smpl
```

```python
# 確認用
pd.pivot_table(base_smpl, index='flg_sts', columns='flg_train', margins=True, aggfunc='count')
```

# データ保存

```python
base_smpl.to_pickle(data_path / ('%s_base_smpl.pkl' % prefix))
```
